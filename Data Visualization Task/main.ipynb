```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axies = np.array([0, 10])
y_axies = np.array([10, 250])

plt.plot(x_axies, y_axies)
```

```python
plt.xlabel("X-Axies")
plt.xlabel("Y-Axies")
plt.title("AI Data Visulization")
plt.plot(x_axies, y_axies)
```

```python
plt.plot(x_axies, y_axies, "o")
```

```python
plt.plot(x_axies, y_axies, "o-r")
```

```python
plt.plot(x_axies, y_axies, "o--r")
```

```python
plt.plot(x_axies, y_axies, "o", ls="--", lw="30.2", c="b")
```

```python
x_axies = np.array([0, 2, 5, 10])
y_axies = np.array([10, 100, 50, 250])

plt.plot(x_axies, y_axies, "o-g")
```

```python
x_axies = np.array([0, 2, 5, 10])
x_axies2 = np.array([2, 3, 5,8])
y_axies = np.array([10, 220, 100, 220])
y_axies2 = np.array([10, 50, 30, 100])


plt.plot(x_axies, y_axies, x_axies2, y_axies2, "o-g")
plt.grid()
plt.savefig("Graph with Grid.png")
```
