```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


dataset = pd.read_csv('insurance.csv')

############################################################
# Finding Duplicated Values using Duplicated() and
# droping the values using dataframe.drop()
############################################################

dup = dataset[dataset.duplicated()].index
print("#"*60 + "\nDuplicated Row....\n" + "#"*60)
print(dup[0])

# Getting the row with duplicated values

print(dataset.iloc[dup])

# Drop the row with duplicated Value
print("#"*60 + "\nDroping The Row 581 ....\n" + "#"*60)
dataset.drop(index=dup, inplace=True)
print(dataset[dataset.columns][579:585])
```

```python
################################################################
# Converting Categorical Data into Numerical Using LabelEncoder
################################################################

# Gender Col to Numeric
label = LabelEncoder()
label.fit(dataset.sex.drop_duplicates())
dataset.sex = label.transform(dataset.sex)

# Smoker Col to Numeric
label.fit(dataset.smoker.drop_duplicates())
dataset.smoker = label.transform(dataset.smoker)

# Region Col to Numeric
label.fit(dataset.region.drop_duplicates())
dataset.region = label.transform(dataset.region)


dataset.head()
```

```python
############################################################
# Splitting Our Dataset into Train and Test Dataset
# Easch Dataset Contain X, and Y for test and Train
# i.e. X_Train, Y_Train, X_Test, Y_Test
############################################################

x = dataset.drop('expenses', axis=1)
y = dataset[['expenses']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
############################################################
# Applying Linear Regression Model on Dataset
############################################################

Linear_Model = LinearRegression()
Linear_Model.fit(x_train, y_train)

predict = Linear_Model.predict(x_test)
predict
```

```python
Linear_Model.score(x_test, predict)
```

```python
for idx, col_name in enumerate(x_train.columns):
    print(f"the Accuracy of {col_name} is: {Linear_Model.coef_[0][idx]}")
```
